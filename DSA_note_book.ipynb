## **Bubble sort implementation of dictionaries by ID**

```python

# Sample student data
students = [
    {'id': 'DBU1601532', 'name': 'Dawit', 'dept': 'DS', 'cgpa': 3.34},
    {'id': 'DBU1609351', 'name': 'Abrham', 'dept': 'CS', 'cgpa': 3.12},
    {'id': 'DBU1602697', 'name': 'Soliana', 'dept': 'IS', 'cgpa': 2.74},
    {'id': 'DBU1602344', 'name': 'Rahel', 'dept': 'CS', 'cgpa': 3.10},
    {'id': 'DBU1609877', 'name': 'Lidia', 'dept': 'SE', 'cgpa': 2.95},
    {'id': 'DBU1605562', 'name': 'Biniam', 'dept': 'IT', 'cgpa': 3.40},
    {'id': 'DBU1609218', 'name': 'Yohannes', 'dept': 'DS', 'cgpa': 2.85},
    {'id': 'DBU1607710', 'name': 'Kidus', 'dept': 'SE', 'cgpa': 3.20},
    {'id': 'DBU1602655', 'name': 'Netsanet', 'dept': 'CS', 'cgpa': 3.00},
    {'id': 'DBU1602879', 'name': 'Mahlet', 'dept': 'IS', 'cgpa': 3.75},
    {'id': 'DBU1604635', 'name': 'hilina', 'dept': 'IT', 'cgpa': 3.55}
]

def bubble_sort_by_id(student_list):
    n = len(student_list)
    for i in range(n):
        for j in range(0, n - i - 1):
            if student_list[j]['id'] > student_list[j + 1]['id']:
                student_list[j], student_list[j + 1] = student_list[j + 1], student_list[j]
    return student_list

sorted_students = bubble_sort_by_id(students.copy())
print(f"{'ID':<12} {'Name':<10} {'Dept':<5} {'CGPA':<4}")
print("-" * 35)
for student in sorted_students:
    print(f"{student['id']:<12} {student['name']:<10} {student['dept']:<5} {student['cgpa']:<.2f}")
```

## **Insertion sort implementation of dictionaries by Name**

```python
students = [
    {'id': 'DBU1601532', 'name': 'Dawit', 'dept': 'DS', 'cgpa': 3.34},
    {'id': 'DBU1609351', 'name': 'Abrham', 'dept': 'CS', 'cgpa': 3.12},
    {'id': 'DBU1602697', 'name': 'Soliana', 'dept': 'IS', 'cgpa': 2.74},
    {'id': 'DBU1602344', 'name': 'Rahel', 'dept': 'CS', 'cgpa': 3.10},
    {'id': 'DBU1609877', 'name': 'Lidia', 'dept': 'SE', 'cgpa': 2.95},
    {'id': 'DBU1605562', 'name': 'Biniam', 'dept': 'IT', 'cgpa': 3.40},
    {'id': 'DBU1609218', 'name': 'Yohannes', 'dept': 'DS', 'cgpa': 2.85},
    {'id': 'DBU1607710', 'name': 'Kidus', 'dept': 'SE', 'cgpa': 3.20},
    {'id': 'DBU1602655', 'name': 'Netsanet', 'dept': 'CS', 'cgpa': 3.00},
    {'id': 'DBU1602879', 'name': 'Mahlet', 'dept': 'IS', 'cgpa': 3.75}
]

def insertion_sort_by_name(student_list):
    for i in range(1, len(student_list)):
        key_student = student_list[i]
        j = i - 1
        while j >= 0 and student_list[j]['name'].lower() > key_student['name'].lower():
            student_list[j + 1] = student_list[j]
            j -= 1
        student_list[j + 1] = key_student
    return student_list


sorted_students = insertion_sort_by_name(students.copy())


print(f"{'ID':<12} {'Name':<10} {'Dept':<5} {'CGPA':<4}")
print("-" * 35)
for student in sorted_students:
    print(f"{student['id']:<12} {student['name']:<10} {student['dept']:<5} {student['cgpa']:<.2f}")
```

## **selection sort implementation of dictionaries by CGPA in Descending order**

```python

students = [
    {'id': 'DBU1601532', 'name': 'Dawit', 'dept': 'DS', 'cgpa': 3.34},
    {'id': 'DBU1609351', 'name': 'Abrham', 'dept': 'CS', 'cgpa': 3.12},
    {'id': 'DBU1602697', 'name': 'Soliana', 'dept': 'IS', 'cgpa': 2.74},
    {'id': 'DBU1602344', 'name': 'Rahel', 'dept': 'CS', 'cgpa': 3.10},
    {'id': 'DBU1609877', 'name': 'Lidia', 'dept': 'SE', 'cgpa': 2.95},
    {'id': 'DBU1605562', 'name': 'Biniam', 'dept': 'IT', 'cgpa': 3.40},
    {'id': 'DBU1609218', 'name': 'Yohannes', 'dept': 'DS', 'cgpa': 2.85},
    {'id': 'DBU1607710', 'name': 'Kidus', 'dept': 'SE', 'cgpa': 3.20},
    {'id': 'DBU1602655', 'name': 'Netsanet', 'dept': 'CS', 'cgpa': 3.00},
    {'id': 'DBU1602879', 'name': 'Mahlet', 'dept': 'IS', 'cgpa': 3.75}
]
# Selection Sort by CGPA (Descending)
def selection_sort_by_cgpa_desc(student_list):
    n = len(student_list)
    for i in range(n):
        max_index = i
        for j in range(i+1, n):
            if student_list[j]['cgpa'] > student_list[max_index]['cgpa']:
                max_index = j
        student_list[i], student_list[max_index] = student_list[max_index], student_list[i]
    return student_list

sorted_students = selection_sort_by_cgpa_desc(students.copy())

print(f"{'ID':<12} {'Name':<10} {'Dept':<5} {'CGPA':<4}")
print("-" * 35)
for student in sorted_students:
    print(f"{student['id']:<12} {student['name']:<10} {student['dept']:<5} {student['cgpa']:<.2f}")
```

## **performance comparison of the three sorting algorithms:**



To compare the performance of the Bubble Sort, Insertion Sort, and Selection Sort algorithms used in this implementation,  analyze their time complexity, number of comparisons/swaps, and practical efficiency for the given dataset.

## ***1. Bubble Sort (Sorted by id)***
#### **Time Complexity:**

$\checkmark$ Worst/Average Case: O(n²)

$\checkmark$ Best Case (Optimized): O(n) (if the list is already sorted)

#### **Key Characteristics:**

$\checkmark$ Repeatedly swaps adjacent elements if they are in the wrong order.

$\checkmark$ The unoptimized version (used in your code) always runs in O(n²), even for sorted lists.

$\checkmark$ Inefficient for large datasets due to excessive swaps.

#### **Performance for 10 Students:**

$\checkmark$ Performs ~45 comparisons for n=10 (since n(n-1)/2 = 10*9/2 = 45).

$\checkmark$ Suitable for small datasets but impractical for large-scale data.

## ***2. Insertion Sort (Sorted by name)***
#### **Time Complexity:**

$\checkmark$ Worst/Average Case: O(n²)

$\checkmark$ Best Case: O(n) (if the list is nearly sorted)

#### **Key Characteristics:**

$\checkmark$ Builds the sorted list one element at a time by inserting each element into its correct position.

$\checkmark$ Efficient for small or partially sorted datasets.

$\checkmark$ Fewer swaps than Bubble Sort in practice.

#### **Performance for 10 Students:**

$\checkmark$ Performs ~25 comparisons on average for random data (better than Bubble Sort).

$\checkmark$ Ideal for small datasets like this one.

## ***3. Selection Sort (Sorted by cgpa in descending order)***
#### **Time Complexity:**

$\checkmark$ Always O(n²) (no best/worst case distinction).

#### **Key Characteristics:**

$\checkmark$ Repeatedly selects the maximum element and swaps it into place.

$\checkmark$ Minimizes swaps (only O(n) swaps total) but performs the same number of comparisons as Bubble Sort.

$\checkmark$ No early termination; always scans the entire unsorted portion.

#### **Performance for 10 Students:**

$\checkmark$ Performs ~45 comparisons (same as Bubble Sort) but only 10 swaps.

$\checkmark$ More swap-efficient than Bubble Sort but still slow for large datasets.



## ***advantages and disadvantages of each algorithm for sorting <center>dictionary data***

## ***1. Bubble Sort***
#### **Advantages:**

$\checkmark$ Simplicity: Easy to implement and understand.
$\checkmark$ Stability: Maintains the order of equal elements (e.g., students with the same id remain in their original order).

$\checkmark$ Early Termination: With optimization (checking if swaps occurred), it can exit early if the list is already sorted.

#### **Disadvantages:**

$\checkmark$ Inefficiency: Worst-case and average-case time complexity is O(n²), making it impractical for large datasets.
For 1,000 students: ~500,000 comparisons/swaps.

$\checkmark$ Unnecessary Swaps: Even in partially sorted data, it repeatedly swaps adjacent elements.

**Best For:**
Small datasets (e.g., 10–50 entries) or educational purposes to demonstrate sorting logic.

## ***2. Insertion Sort***
#### **Advantages:**

$\checkmark$ Adaptive: Efficient for small or nearly sorted datasets (best-case time complexity O(n)).
Example: Adding a new student to an already sorted list.

$\checkmark$ Stability: Preserves the original order of equal keys (e.g., students with the same name).

$\checkmark$ In-Place Sorting: Requires no additional memory.

$\checkmark$ Intuitive Logic: Mimics how humans sort cards in their hands.

#### **Disadvantages:**

$\checkmark$ Quadratic Time: Worst-case time complexity O(n²) for reverse-sorted data.
For 1,000 students: ~500,000 comparisons/shifts.

$\checkmark$ Shifting Overhead: Requires shifting elements to insert a value, which can be slow for large datasets.

**Best For:**
Small datasets, incremental updates (e.g., adding new students to a sorted list), or partially sorted data.

## ***3. Selection Sort***
#### **Advantages:**

$\checkmark$ Swap Efficiency: Performs only O(n) swaps (minimal compared to Bubble/Insertion Sort).
Useful if swapping large dictionaries is expensive (e.g., dictionaries with many keys).

$\checkmark$ Simplicity: Easy to implement for sorting by a specific key (e.g., cgpa).

$\checkmark$ In-Place Sorting: No extra memory required.

#### **Disadvantages:**

$\checkmark$ Fixed Comparisons: Always requires O(n²) comparisons, even for sorted data.
For 1,000 students: ~500,000 comparisons.

$\checkmark$ Instability: Does not preserve the order of equal elements (e.g., students with the same cgpa may swap positions).

$\checkmark$ No Early Termination: Always scans the entire unsorted portion.

**Best For:**
Small datasets where minimizing swaps is critical (rarely used in practice for dictionary data).

##### **Key Considerations for Dictionary Data**
$\circ$ **Comparison Overhead:**

Accessing dictionary keys (e.g., student['id']) adds slight overhead compared to sorting simple lists.

All three algorithms handle this similarly, as they compare values via key lookups.

$\circ$ **Stability:**

Critical if sorting by multiple criteria (e.g., first by dept, then by cgpa).

Bubble Sort and Insertion Sort are stable; Selection Sort is not.

$\circ$ **Data Size:**

For large datasets (1,000+ entries), all three algorithms are inefficient. Use O(n log n) algorithms like Merge Sort or Quick Sort instead.
